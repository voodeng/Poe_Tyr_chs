```python
def getSpellsDesc(pageUrl):
    try:
        content = get_page(pageUrl)
        pattern = re.compile('<table class="quote.*?>.*?<i>(.*?)</i>.*?</table>',re.S)
        result = re.search(pattern,content)
        str = re.sub(r'</?\w+[^>]*>','',result.group(1))
        return str
    except:
        return ''
```
