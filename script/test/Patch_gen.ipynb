```python
patch_result = os.path.join(STORAGE_DIR, u"patch.xlsx")
patched_result = os.path.join(STORAGE_DIR, u"new_patch.xlsx")


# 神祇，名词，人名，地名
# _sn = '人名'
# pr = pd.read_excel(patch_result,sheetname=_sn).fillna('')
# pr = pr.copy()

# xlrd 读取表名
# book = xlrd.open_workbook(patch_result) 
# for sheet in book.sheets():
#         print( sheet.name)

# openyxl
# book = load_workbook(patch_result)
# for sheet in book:
#     print(sheet.title)

tr_list = {}

def handler_patch(list):
    for i,row in list:
        
        item = {}
        rt = row['Eng']

        if rt == '':
            continue

        rt = rt.split(' - ')
        if len(rt) > 2:
            rt[0] = "-".join([rt[0], rt[1]])
            rt[1] = rt[2]
        if len(rt) > 1:
            item['Eng'] = rt[0].strip()
            if rt[1] == '':
                continue
            item['Chs'] = fix_text(rt[1].strip())
        else:
            item['Eng'] = row['Eng'].strip().replace('- ','')
            if row['Chs'] == '':
                continue
            item['Chs'] = fix_text(row['Chs'].strip())
        if len(row)>2 is not None:
            item['Rep'] = row[2].strip()
        else:
            item['Rep'] =''
            
        tr_list[_sn].append(item)
#         print(item)

book = openpyxl.load_workbook(patch_result)
# workbook = load_workbook(filename=filename)
# sheets = workbook.get_sheet_names()
# sheet = workbook.get_sheet_by_name(sheets[0])
for sheet in book:
    print(sheet.title)
    _sn = sheet.title
    pr = pd.read_excel(patch_result,sheetname=_sn).fillna('')
    tr_list[_sn] = []
    handler_patch(pr.iterrows())
    
```
